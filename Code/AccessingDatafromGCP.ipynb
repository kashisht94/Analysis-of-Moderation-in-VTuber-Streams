```python
from google.colab import drive
from azure.storage.blob import BlobServiceClient
import pandas as pd
import io
```

```python
drive.mount('/content/drive')
```

```python
config_path =  r'/content/drive/MyDrive/Sem2/BigData/config.txt'
```

```python
config = {}

with open(config_path, 'r') as file:
    for line in file:
        key, value = line.strip().split(': ')
        config[key] = value

#account_name = config.get('accountName')
#storage_name = config.get('storageName')
#access_key = config.get('accessKey')
connection_string = config.get('connectionString')
```

# Pandas Method

```python
blob_service_client = BlobServiceClient.from_connection_string(connection_string)
print('connected succesfully!')
container_client = blob_service_client.get_container_client("vtubercontainer")
```

```python
# List blobs in the container
blob_list = container_client.list_blobs()
for blob in blob_list:
    print(blob.name)
```

```python
ban_blob = "ban_events.parquet"
blob_client = container_client.get_blob_client(ban_blob)
ban_blob_data = blob_client.download_blob()
```

```python
ban_df = pd.read_parquet(io.BytesIO(ban_blob_data.readall()))
```
